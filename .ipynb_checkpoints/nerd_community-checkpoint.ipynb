## Clean and Tokenize data : 

```python
print len(a)
```

```python
from nltk.corpus import stopwords
```

```python
stopword_list = set(stopwords.words("english"))
stopword_list.update(['.', ',', '"', "'", '?', '!', ':', ';', '(', ')', '[', ']', '{','}']) 
```

```python
all_tokens = []
for sentence in a:
    tokens = nltk.word_tokenize(sentence)
    for t in tokens:
        if t not in stopword_list: # Remove token if token is in stop word list
            all_tokens.append(t)
```

```python
len(all_tokens)
```

## What hashtags do students frequently use? : 

Students tend to use many hashtags for tagging/categorizing their posts. Here we take a look at those hashtags.

```python
from nltk.tokenize import TweetTokenizer
```

```python
tknzr = TweetTokenizer()
tweet_tokens = [] 
for sentence in a:
    tokens = tknzr.tokenize(sentence)
    for t in tokens:
        if t not in stopword_list: # Remove token if token is in stop word list
            tweet_tokens.append(t)
```

```python
hashtags = []
for token in tweet_tokens:
    if token.startswith("#") and token != '#':
        hashtags.append(token.lower())
```

```python
len(hashtags)
```

```python
fdist_hashtag2 = nltk.FreqDist(hashtags)
fdist_hashtag2.most_common(30)
```

```python
fdist_hashtag = nltk.FreqDist(hashtags)
```

```python
fdist_hashtag.most_common(30)
```

```python
fdist_hashtag.plot(30, cumulative = False)
```

## What are the topics? (Using LDA)

### Preprocess tokens: 

https://rstudio-pubs-static.s3.amazonaws.com/79360_850b2a69980c4488b1db95987a24867a.html
https://www.youtube.com/watch?v=BuMu-bdoVrU

```python
from nltk.tokenize import RegexpTokenizer

from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim

tokenizer = RegexpTokenizer(r'\w+')


# Create p_stemmer of class PorterStemmer
p_stemmer = PorterStemmer()
    
# list for tokenized documents in loop
texts = []

# loop through document list
for i in a:
    
    # clean and tokenize document string
    raw = i.lower()
    tokens = tokenizer.tokenize(raw)

    # remove stop words from tokens
    stopped_tokens = [i for i in tokens if not i in stopword_list]
    
    # stem tokens
    stemmed_tokens = [p_stemmer.stem(i) for i in stopped_tokens]
    
    # add tokens to list
    texts.append(stemmed_tokens)

# turn our tokenized documents into a id <-> term dictionary
dictionary = corpora.Dictionary(texts)
    
# convert tokenized documents into a document-term matrix
corpus = [dictionary.doc2bow(text) for text in texts]

# generate LDA model
ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics=5, id2word = dictionary, passes=20)
```

```python
print(ldamodel.print_topics(num_topics=5, num_words=5))
```
